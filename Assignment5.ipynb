```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic Ship Passengers dataset
titanic_df = pd.read_csv("/content/titanic_train.csv")
```

```python
# Check data types of each feature
print(titanic_df.dtypes)
```

```python
# Generate a summary of the dataset
print(titanic_df.describe())
```

```python
# List the names of columns/features
print(titanic_df.columns)
```

```python
# Plot numeric features
sns.pairplot(titanic_df, vars=['Age', 'Fare'])

# Check relative size of survived/un-survived
sns.countplot(x='Survived', data=titanic_df)

# Check if any pattern on gender
sns.countplot(x='Sex', hue='Survived', data=titanic_df)

# Check if any pattern on passenger class
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)

# Check class-wise survival rate
sns.barplot(x='Pclass', y='Survived', data=titanic_df)

# Check sibilings
sns.countplot(x='SibSp', hue='Survived', data=titanic_df)

# Overall age distribution
sns.histplot(titanic_df['Age'], kde=True)

# Class-wise age distribution
sns.boxplot(x='Pclass', y='Age', data=titanic_df)
```
