```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report

# Load the classified dataset into a dataframe using pandas
df = pd.read_csv('/content/Classified_Data.txt')

# Check the data types of each feature(column) in the dataset
print(df.dtypes)

# Generate a summary of the dataset for min, max, stddev, quartile values for 25%, 50%, 75%, 90%
print(df.describe())

# List the names of columns/features in the dataset
print(df.columns)

# Scale the features using StandardScaler and transform the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Split the data into X and Y
X = df_scaled[:, :-1]
Y = df_scaled[:, -1]

# Split the data into training and testing sets using train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Apply the KNN Classifier model
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the data to the Classifier Model using fit
knn.fit(X_train, Y_train)
```

```python
# Generate the confusion matrix to estimate the correction among features
Y_pred = knn.predict(X_test)
conf_mat = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:")
print(conf_mat)

# Generate the classification report using classification report
print("Classification Report:")
print(classification_report(Y_test, Y_pred))
```
